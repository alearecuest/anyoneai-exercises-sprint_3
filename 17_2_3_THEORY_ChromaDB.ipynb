# **Understanding Vector Databases with ChromaDB**

## **Introduction to Vector Databases**

A **Vector Database** is designed to efficiently store, index, and search vectors—mathematical representations of data that capture semantic meaning. In many AI and machine learning applications, data such as text, images, and audio are converted into numerical vectors using techniques like embeddings. These vectors preserve the relationships between data points based on their meaning or content.

### **Why Use a Vector Database?**
1. **Efficient Storage**: Traditional databases are not optimized for storing and searching high-dimensional vectors, while vector databases are designed specifically for this purpose.
2. **Fast Similarity Search**: They allow for rapid retrieval of similar items, often using algorithms like **k-nearest neighbors (k-NN)** to find vectors that are semantically close to a query vector.
3. **Scalability**: Vector databases are capable of handling millions or even billions of vectors, making them suitable for large-scale machine learning applications such as recommendation systems, search engines, and more.

## **ChromaDB**

**ChromaDB** is a modern vector database designed for AI applications. It provides high-performance vector search capabilities and can be integrated easily with machine learning pipelines.

### **Example Use Cases of ChromaDB**
- Storing sentence embeddings for document retrieval
- Building a recommendation engine based on product embeddings
- Image similarity search using image embeddings

```python
# Importing necessary libraries
import chromadb
from chromadb.config import Settings
from chromadb.utils import embedding_functions

# For embedding generation using SentenceTransformers
from sentence_transformers import SentenceTransformer
```

## **Creating a Simple ChromaDB Collection**

ChromaDB allows you to create collections that store vectors (embeddings). Below is an example of creating a collection and adding some text data to it.


```python
Settings()
```

```python
# Initialize ChromaDB client
client = chromadb.Client(Settings())

# to persist the DB on disk (otherwise it just remains on RAM memory)
client = chromadb.PersistentClient(path="./sentence_db")

# Create a collection
collection = client.create_collection("my_text_collection")
```

## **Generating Embeddings**

We'll use a pre-trained model from the **SentenceTransformers** library to generate embeddings for text. These embeddings are high-dimensional vectors that represent the semantic meaning of the text.


```python
# Load the SentenceTransformer model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Example text data
documents = ["This is a great book about AI.",
             "ChromaDB is a vector database that implements fast algorithms",
             "Vector databases are highly useful for AI-driven applications.",
             "The weather is nice today.",
             "Apples are red.",
             "Artificial intelligence is revolutionizing industries by automating tasks and enabling advanced data analysis.",
             "The sound of waves crashing against the shore creates a sense of peace that only nature can provide.",
             "Quantum mechanics challenges our understanding of the universe by revealing how particles behave in strange, unpredictable ways.",
             "The fall of the Roman Empire marked a significant turning point in the history of Western civilization.",
             "NASA is preparing for its next mission to Mars, aiming to explore the planet's surface for signs of ancient life.",
             "Existentialism explores the meaning of life, encouraging individuals to define their own purpose in a seemingly indifferent universe.",
             "Vincent van Gogh's use of bold colors and expressive brushstrokes has made his work timeless and instantly recognizable.",
             "Classical symphonies evoke powerful emotions through the intricate interplay of instruments and dynamic shifts in melody.",
             "The final match of the World Cup was a nail-biting contest, with both teams giving their all until the last whistle.",
             "Exploring the narrow streets of a bustling market in Marrakech is a sensory experience like no other.",
             "Maintaining a balanced diet and regular exercise are key components of a healthy lifestyle.",
             "Online learning platforms have transformed education by making high-quality courses accessible to learners around the globe.",
             "In 'Pride and Prejudice,' Jane Austen masterfully explores the complexities of relationships and societal expectations.",
             "Entrepreneurship requires a blend of creativity, risk-taking, and perseverance to turn innovative ideas into successful ventures.",
             "Cognitive behavioral therapy helps individuals identify and challenge negative thought patterns to improve mental well-being."
]
```

```python
# Generate embeddings
embeddings = model.encode(documents).tolist()

# Adding documents and their embeddings to the ChromaDB collection
for i, doc in enumerate(documents):
    collection.add(ids=[str(i)], documents=[doc], embeddings=[embeddings[i]])
```

```python
print(len(embeddings))
print(len(documents))
print(embeddings[0])
print(len(embeddings[0]))
```

## **Querying the Vector Database**

Now that we have stored embeddings in the ChromaDB collection, we can perform similarity searches. Given a query, the database will return the most semantically similar documents.


```python
import pandas as pd

# Query text
# query = "Storing vectors in a database."
query = "science"

# Generate embedding for the query
query_embedding = model.encode([query]).tolist()

# Perform similarity search
results = collection.query(query_embeddings=query_embedding, n_results=5)

# Display results
results_list = []
for i, score in enumerate(results['distances'][0]):
    results_list.append([score, results['documents'][0][i]])

pd.set_option('display.max_colwidth', None)
pd.DataFrame(results_list, columns=['score', 'document'])
```

## **Deleting and Updating Records**

ChromaDB also allows you to delete and update records within your collection.


```python
# Delete a document by ID
collection.delete(ids=["0"])

# Update a document
collection.update(ids=["1"], documents=["Updated document about machine learning."])

# Check updated results
updated_results = collection.query(query_embeddings=query_embedding, n_results=4)

updated_results_list = []
for i, score in enumerate(updated_results['distances'][0]):
    updated_results_list.append([score, updated_results['documents'][0][i]])

pd.set_option('display.max_colwidth', None)
pd.DataFrame(updated_results_list, columns=['score', 'document'])
```
