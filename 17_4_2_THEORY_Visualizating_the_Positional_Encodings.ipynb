```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf

def positional_encoding(position, d_model):
    # Create a matrix of angles
    angle_rads = get_angles(np.arange(position)[:, np.newaxis], # column vector of positions (0, pos)
                            np.arange(d_model)[np.newaxis, :], # row vector of dimensions (0, d_model)
                            d_model # Embedding
                            )

    # Apply sine and cosine to the angles
    # Even values are assigned to sin
    angle_rads[:, 0::2] = np.sin(angle_rads[:, 0::2])
    # Odd values are asigned to cos
    angle_rads[:, 1::2] = np.cos(angle_rads[:, 1::2])

    # Convert to tensor
    pos_encoding = angle_rads[np.newaxis, ...]
    return tf.cast(pos_encoding, dtype=tf.float32)

def get_angles(pos, i, d_model):
    # The angles are defined by the equation
    # pos = position
    # i = depth
    # d_model = model dimension
    # angle = pos / 10000^(2(i)/d_model)
    angle_rates = 1 / np.power(10000, (2 * (i // 2)) / np.float32(d_model))
    return pos * angle_rates
```

```python
# Parameters
position = 1
d_model = 512

# Generate positional encodings
pos_encoding = positional_encoding(position, d_model)

# Convert to numpy array for plotting
pos_encoding = pos_encoding.numpy()

# Plot the positional encodings
plt.figure(figsize=(10, 8))
plt.pcolormesh(pos_encoding[0], cmap='viridis')
plt.xlabel('Depth')
plt.xlim((0, d_model))
plt.ylabel('Position')
plt.ylim((0, position))
plt.colorbar()
plt.title("Positional Encoding")
plt.show()
```
