## 1) Natural Language Toolkit

The Natural Language Toolkit (NLTK) is a powerful Python library used for working with human language data (text). It provides a wide range of tools for natural language processing (NLP) tasks including:

- Text Processing Utilities: NLTK includes modules for tasks like tokenization (splitting text into words or sentences), stemming (reducing words to their root forms), and part-of-speech tagging (labeling words as nouns, verbs, etc.).

- Corpora and Lexical Resources: It offers access to large collections of text (corpora) and lexical resources like WordNet, which helps in working with word meanings and relationships.

- Support for Advanced NLP Tasks: NLTK also provides tools for more complex tasks, such as named entity recognition (finding names of people or places in text), parsing (analyzing sentence structure), and text classification. ## 3) Stemming

Stemming is a text normalization technique in Natural Language Processing (NLP) that reduces words to their root or base form, called a "stem." The goal of stemming is to remove suffixes and prefixes from words, simplifying them so that different variations of a word (like "carrying," "carried," and "carry") are treated as the same term. This process helps in tasks such as search engines or text analysis by improving efficiency and accuracy when working with different word forms.

In the code provided, we use two popular stemming algorithms: the Porter Stemmer and the Snowball Stemmer from the NLTK library. Both algorithms help reduce the words in the list to their stems, but they use slightly different rules for handling suffixes and word forms. The output will show how each stemmer handles the same words.

```python
# Import stemming library
from nltk.stem import PorterStemmer
porter = PorterStemmer()

# Word-list for stemming
word_list = ["carry","carrying","carried"]
for w in word_list:
    print(porter.stem(w))
```

## 4) Lemmatization

Lemmatization is a natural language processing (NLP) technique that reduces words to their base or dictionary form, known as a lemma. Unlike stemming, which simply cuts off prefixes and suffixes, lemmatization considers the context and meaning of the word to return a valid root form. This makes lemmatization more accurate in preserving the actual meaning of words across different inflections, such as verb tenses or plural nouns.

In the provided code, the WordNetLemmatizer from the NLTK library is used to demonstrate lemmatization. You can specify the part of speech (POS)—such as **verb (v), noun (n), adjective (a), or adverb (r)**—to guide the lemmatizer in accurately determining the correct base form of the word. The examples show how lemmatization can return different results depending on the part of speech.


```python
from nltk import WordNetLemmatizer

lemma = WordNetLemmatizer()
word_list = ["carry","carrying","carried"]

for w in word_list:
    print(lemma.lemmatize(w, pos="v"))
```

```python
for w in word_list:
    print(lemma.lemmatize(w))
```

```python
from nltk.stem import WordNetLemmatizer

# POS: Part of speech
lemmatizer = WordNetLemmatizer()

print(lemmatizer.lemmatize('reading', pos="v")) # Verb
print(lemmatizer.lemmatize('reading', pos="n")) # Noun
print(lemmatizer.lemmatize('reading', pos="a")) # Adjective
print(lemmatizer.lemmatize('reading', pos="r")) # Adverb
```

## 6) WordNet

WordNet is a large lexical database of English that groups words into sets of synonyms called synsets. It is widely used in natural language processing (NLP) for tasks involving word meanings, relationships, and language understanding. Each synset in WordNet represents a specific concept, and words within a synset share similar meanings. Additionally, WordNet provides definitions, example sentences, and various relationships between words, such as synonyms, antonyms, and hypernyms (more general terms).

In the provided code, the WordNet module from the NLTK library is used to explore different aspects of word meanings:

1. The first part retrieves definitions for words like "Sun" and "Fun" from WordNet synsets.
2. Synonyms for words are obtained by accessing the lemmas within each synset.
3. Antonyms (opposite meanings) are found by exploring the antonyms associated with the word's lemmas.
4. Word similarity is measured using the Wu-Palmer similarity, which compares two words based on their meanings in the WordNet hierarchy.

```python
# Import wordnet
# nltk.download('wordnet')
from nltk.corpus import wordnet

for words in wordnet.synsets("Sun"):
    print(words.definition())
```

```python
# Word meaning with definitions

# LEMMA - a heading indicating the subject or argument of a literary composition or annotation

for words in wordnet.synsets("Fun"):
    print(words.name())
    print(words.definition())
    print(words.examples())
    for lemma in words.lemmas(): # Corresponding synonyms
        print(lemma)
    print("-------------")
```

```python
# Finding synonyms

# Empty list to store synonyms
synonyms = []

for words in wordnet.synsets('Fun'):
    for lemma in words.lemmas():
        synonyms.append(lemma.name())

set(synonyms)
```

```python
word_list = ["carry","carrying","carried"]
wordnet_word = []

for word in word_list:
  w = wordnet.synsets(word)[0]
  wordnet_word.append(w)
```

```python
wordnet_word[0].lemmas()[0].name()
```

```python
wordnet_word[1].lemmas()[0].name()
```

```python
# Finding antonyms

# Empty list to store antonyms
antonyms = []

for words in wordnet.synsets('Up'):
    for lemma in words.lemmas():
        if lemma.antonyms():
            antonyms.append(lemma.antonyms()[0].name())

# Print antonyms
set(antonyms)
```

```python
# Similarity in words
word1 = wordnet.synsets("car","n")[0]
word2 = wordnet.synsets("vehicle","n")[0]

# Check similarity
print(word1.lemmas()[0].name())
print(word2.lemmas()[0].name())
print(word1.wup_similarity(word2))
```

## 7) Integer_Encoding and One_Hot_Encoding

Integer Encoding and One-Hot Encoding are common techniques used in machine learning to convert categorical data (like labels or text) into numerical representations that models can work with.

Integer Encoding: This method involves assigning a unique integer to each category. For example, in the given data ['cold', 'cold', 'warm', 'cold', 'hot', 'hot'], "cold" might be encoded as 0, "warm" as 1, and "hot" as 2. This method is straightforward but can introduce unintended relationships between the categories since the numbers have an implicit order.

One-Hot Encoding: To avoid introducing an order or hierarchy, one-hot encoding is used. In this technique, each category is represented by a binary vector where only one element is "1" (indicating the presence of a specific category) and all others are "0". For example, "cold" might become [1, 0, 0], "warm" becomes [0, 1, 0], and "hot" becomes [0, 0, 1]. This approach ensures that no ordinal relationship is implied between categories.

In the provided code, we first use LabelEncoder to perform integer encoding on the data, followed by OneHotEncoder to transform the integer-encoded data into a binary matrix for one-hot encoding.

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder

data = ['cold', 'cold', 'warm', 'cold', 'hot', 'hot']
print(data)

# Integer encode
label_encoder = LabelEncoder()
integer_encoded = label_encoder.fit_transform(data)
print(integer_encoded)

# Binary encode
onehot_encoder = OneHotEncoder(sparse=False) # No hashing
integer_encoded = integer_encoded.reshape(len(integer_encoded), 1) # Make it column matrix 6x1
onehot_encoded = onehot_encoder.fit_transform(integer_encoded)
print(onehot_encoded)
```

## 8) Bag of Words

The Bag of Words (BoW) model is a common technique used in Natural Language Processing (NLP) to convert text data into numerical features for machine learning algorithms. In this model, a document is represented as a collection of the words it contains, without considering grammar or the order of words. The primary idea is to create a "vocabulary" of all the unique words in a dataset and then represent each document as a vector where each element corresponds to the frequency of a specific word from the vocabulary in that document.

It works in 2 steps:

1. **Vocabulary Creation:** The model first identifies all unique words across the dataset (ignoring duplicates).

2. **Word Frequency:** Each document is then represented as a vector where the value in each position indicates the number of times the corresponding word appears in that document. This forms the "bag of words" for each document, where order is irrelevant.

In the provided code, the CountVectorizer from the Scikit-learn library is used to implement the Bag of Words model. The first part of the code demonstrates how a set of sentences is converted into a matrix where each row corresponds to a sentence and each column corresponds to a unique word in the vocabulary. The second part adds more complexity by using n-grams, which capture sequences of 1 or 2 words (instead of individual words), providing more context in the vectorized output. This method is useful for transforming text into a form that can be used in machine learning models for tasks like text classification, sentiment analysis, or information retrieval.

```python
from sklearn.feature_extraction.text import CountVectorizer

sentences = ['Shyam is a bad boy','Ram is a good boy']
cv = CountVectorizer()                              # Create an object
B_O_W = cv.fit_transform(sentences).toarray()       # Generating output for Bag of Words
print(cv.vocabulary_)
print(cv.get_feature_names_out())
print('Bag of Words')
print(B_O_W)
```

```python
sentences = ['Shyam is a good boy','Ram is not a good boy']
cv = CountVectorizer(ngram_range=(1, 2))             # Create an object
B_O_W = cv.fit_transform(sentences).toarray()       # Generating output for Bag of Words
print(cv.vocabulary_)
print(cv.get_feature_names_out())
print('Bag of Words')
print(B_O_W)
```

## 9) TF-IDF (Term Frequency Inverse Document Frequency)

TF-IDF (Term Frequency-Inverse Document Frequency) is a statistical measure used in Natural Language Processing (NLP) to evaluate how important a word is to a document in a collection or corpus. It helps reduce the impact of common words (like "is" or "the") and highlight more relevant words that are unique to a document.

- Term Frequency (TF) measures how frequently a word appears in a document.
- Inverse Document Frequency (IDF) reduces the weight of words that appear frequently across many documents, giving more importance to words that are unique or less common in the overall corpus.

The combination of TF and IDF helps balance word frequency and relevance, making TF-IDF a popular technique for tasks like text classification and information retrieval.

In the provided code, the TfidfVectorizer from Scikit-learn is used to calculate the TF-IDF values for a set of sentences. The result is a matrix where each row represents a sentence and each column corresponds to the TF-IDF score of a word, showing how important each word is within that sentence and across the entire set. This transformation is useful for extracting meaningful features from text for further analysis or machine learning tasks.

For a more detailed notebook on TF-IDF, check the Theory Notebook "17.1.2 - THEORY - Implementing TF-IDF" available here: https://drive.google.com/file/d/17HETfTpP__Bdce7qULUKCa5K3isoWe2l/view?usp=sharing

```python
# Import required libraries
from sklearn.feature_extraction.text import TfidfVectorizer

# Sentences for analysis
sentences = ['It is a good book', 'It is not a good book', 'It is a very good book', 'It is very very good book']

# Create an object
vectorizer = TfidfVectorizer(norm=None)

# Generating output for TF_IDF
X = vectorizer.fit_transform(sentences).toarray()

# Total words with their index in model
print(vectorizer.vocabulary_)
print(vectorizer.get_feature_names_out())

# Show the output
print(X)
```
